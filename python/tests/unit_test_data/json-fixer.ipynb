```python
import json

with open('./resample_tests.json', 'r') as file:
    before = json.load(file)
```

```python
def update_dict(dictionary, key, value):
    if value is not None:
        dictionary[key] = value
```

```python
after = {}
for i in before.keys(): # i is test class
    if i == "__SharedData":
        continue
    after[i] = {}
    for j in before[i].keys(): # j is test method
        after[i][j] = {}
        for k in before[i][j].keys(): # input, expected, etc.
            tsdf = {}
            update_dict(tsdf, "ts_col", before[i][j][k].get("ts_col", None))
            update_dict(tsdf, "other_ts_cols", before[i][j][k].get("other_ts_cols", None))
            update_dict(tsdf, "partition_cols", before[i][j][k].get("partition_cols", None))
            update_dict(tsdf,  "sequence_col", before[i][j][k].get("sequence_col", None))
            update_dict(tsdf, "start_ts", before[i][j][k].get("start_ts", None))
            update_dict(tsdf, "end_ts", before[i][j][k].get("end_ts", None))
            update_dict(tsdf, "series", before[i][j][k].get("series", None))
            sdf = {}
            update_dict(sdf, "schema", before[i][j][k].get("schema", None))
            update_dict(sdf, "ts_convert", before[i][j][k].get("ts_convert", None))
            update_dict(sdf, "data", before[i][j][k].get("data", None))
            after[i][j][k] = {
                        "tsdf": tsdf,
                        "df": sdf,
                        "$ref": before[i][j][k].get("$ref", None)
                    }
```

```python
after_2 = {}
for i in before.keys(): # i is test class
    if i != "__SharedData":
        continue
    after_2[i] = {}
    for j in before[i].keys(): # j is test method
        tsdf = {}
        update_dict(tsdf, "ts_col", before[i][j].get("ts_col", None))
        update_dict(tsdf, "other_ts_cols", before[i][j].get("other_ts_cols", None))
        update_dict(tsdf, "partition_cols", before[i][j].get("partition_cols", None))
        update_dict(tsdf,  "sequence_col", before[i][j].get("sequence_col", None))
        update_dict(tsdf, "start_ts", before[i][j].get("start_ts", None))
        update_dict(tsdf, "end_ts", before[i][j].get("end_ts", None))
        update_dict(tsdf, "series", before[i][j].get("series", None))
        sdf = {}
        update_dict(sdf, "schema", before[i][j].get("schema", None))
        update_dict(sdf, "ts_convert", before[i][j].get("ts_convert", None))
        update_dict(sdf, "data", before[i][j].get("data", None))
        after_2[i][j] = {
                    "tsdf": tsdf,
                    "df": sdf,
                }
```
