```python
import json

with open('./resample_tests.json', 'r') as file:
    before = json.load(file)
```

```python
after = {}
for i in before.keys(): # i is test class
    if i == "__SharedData":
        continue
    after[i] = {}
    for j in before[i].keys(): # j is test method
        after[i][j] = {}
        for k in before[i][j].keys(): # input, expected, etc.

            after[i][j][k] = {
                        "tsdf": {
                            "ts_col": before[i][j][k].get("ts_col", None),
                            "other_ts_cols": before[i][j][k].get("other_ts_cols", None),
                            "partition_cols": before[i][j][k].get("partition_col", None),
                            "sequenc_col": before[i][j][k].get("sequence_col", None),
                            "start_ts": before[i][j][k].get("start_ts", None),
                            "end_ts": before[i][j][k].get("end_ts", None),
                            "series": before[i][j][k].get("series", None),
                            
                        },
                        "df": {
                            "schema": before[i][j][k].get("schema", None),
                            "ts_convert": before[i][j][k].get("ts_convert", None),
                            "data": before[i][j][k].get("data", None)
                        },
                        "$ref": before[i][j][k].get("$ref", None)
                    }
```

```python
after_2 = {}
for i in before.keys(): # i is test class
    if i != "__SharedData":
        continue
    after_2[i] = {}
    for j in before[i].keys(): # j is test method
            after_2[i][j] = {
                        "tsdf": {
                            "ts_col": before[i][j].get("ts_col", None),
                            "other_ts_cols": before[i][j].get("other_ts_cols", None),
                            "partition_cols": before[i][j].get("partition_col", None),
                            "sequence_col": before[i][j].get("sequence_col", None),
                            "start_ts": before[i][j].get("start_ts", None),
                            "end_ts": before[i][j].get("end_ts", None),
                            "series": before[i][j].get("series", None),
                            
                        },
                        "df": {
                            "schema": before[i][j].get("schema", None),
                            "ts_convert": before[i][j].get("ts_convert", None),
                            "data": before[i][j].get("data", None)
                        },
                        "$ref": before[i][j].get("$ref", None)
                    }
```

```python
combined = after | after_2
```
